# Data preprocessing

```python
df_required=df.drop('Car_Name', axis=1) #axis =1 means it is denoting the column and axis = 0 means its denoting row.
```

```python
df_required
```

```python
df_required['Present_year']=2023
```

```python
df_required.head()
```

```python
df_required['Age_year']=df_required['Present_year']-df_required['Year']
```

```python
df_required.head()
```

```python
df_required.drop(['Year','Present_year'],axis=1,inplace=True)
```

```python
df_required.head()
```

```python
df_required=pd.get_dummies(df_required)
df_required.head()
```

# Multi collinearity 

```python
df_required_x=df_required.drop('Selling_Price',axis=1)
df_required_x.head()
```

```python
plt.figure(figsize=(12,10))
sn.heatmap(df_required_x.corr(), annot=True)
```
